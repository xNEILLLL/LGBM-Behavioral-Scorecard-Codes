```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('/Users/queencyangela/Desktop/M AMF/Second Sem/MATH 298.1/ILS_home_final.csv')  # Replace with your actual file name

# Display the first few rows
print(df)
```

```python
import pandas as pd
import re
import calendar

# Dictionary to map month abbreviations to numbers
month_dict = {
    "JAN": 1, "FEB": 2, "MAR": 3, "APR": 4, "MAY": 5, "JUN": 6,
    "JUL": 7, "AUG": 8, "SEP": 9, "OCT": 10, "NOV": 11, "DEC": 12
}

# Function to convert "WORST_DELINQUENCY JAN 2024" -> "01/31/2024"
def convert_to_last_day(column_name):
    match = re.search(r'Worst_Del (\w{3}) (\d{4})', column_name, re.IGNORECASE)
    if match:
        month_abbr, year = match.groups()
        if month_abbr.upper() in month_dict:
            month = month_dict[month_abbr.upper()]
            year = int(year)
            last_day = calendar.monthrange(year, month)[1]  # Get last day of the month
            return f"{month:02d}/{last_day:02d}/{year}"  # Convert to MM/DD/YYYY
    return column_name  # If no match, return as is

# Apply renaming function
df.rename(columns={col: convert_to_last_day(col) for col in df.columns}, inplace=True)

df = df.drop(columns=[col for col in df.columns if "DEL1" in col])

# Check the updated column names
print(df)
```
