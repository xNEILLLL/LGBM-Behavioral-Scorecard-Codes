# REMOVE INDETERMINATE

```python
# Convert DELQ_AGE_ACTUAL to numeric to avoid issues
df['DELQ_AGE_ACTUAL'] = pd.to_numeric(df['DELQ_AGE_ACTUAL'], errors='coerce')

# Filter for only M1 to M12 rows before checking DELQ_AGE_ACTUAL
# df_m1_m12 = df[df['MONTH'].isin(['M13', 'M14', 'M15', 'M16', 'M17', 'M18', 'M19', 'M20', 'M21', 'M22', 'M23', 'M24'])]
# df_m1_m12 = df[df['MONTH'].isin(['M73', 'M74', 'M75', 'M76', 'M77', 'M78', 'M79', 'M80', 'M81', 'M82', 'M83', 'M84'])]
df_m1_m12 = df[df['MONTH'].isin(['M73', 'M74', 'M75', 'M76', 'M77', 'M78', 'M79', 'M80', 'M81', 'M82', 'M83', 'M84'])]

# Identify accounts that have at least one row with DELQ_AGE_ACTUAL between 31 and 90 (exclusive)
accounts_to_remove = df_m1_m12.loc[(df_m1_m12['DELQ_AGE_ACTUAL'] > 15) & (df_m1_m12['DELQ_AGE_ACTUAL'] < 92), 'ILS_ACCT_NUM'].unique()

# Keep only accounts that are NOT in accounts_to_remove
df = df[~df['ILS_ACCT_NUM'].isin(accounts_to_remove)]

df.head(20)
```

## BUILD FEATURES

```python
# Ensure relevant columns are numeric
df['DELQ_AGE_ACTUAL'] = pd.to_numeric(df['DELQ_AGE_ACTUAL'], errors='coerce')
df['WORST_DELINQUENCY'] = pd.to_numeric(df['WORST_DELINQUENCY'], errors='coerce')
df['ACTUAL_AMORT_RATIO'] = pd.to_numeric(df['ACTUAL_AMORT_RATIO'], errors='coerce')
df['REMAINING_LOAN_RATIO'] = pd.to_numeric(df['REMAINING_LOAN_RATIO'], errors='coerce')

# Extract unique ILS_ACCT_NUM, ORIG_BOOK_DATE, and MATURITY
df_unique_base = df[['ILS_ACCT_NUM', 'ORIG_BOOK_DATE', 'MATURITY']].drop_duplicates()

# Define month groups for M1-M12 only
month_groups_dict = {
    # 'M1_M12': {
    #     '3mos': ['M10', 'M11', 'M12'],
    #     '6mos': ['M7', 'M8', 'M9', 'M10', 'M11', 'M12'],
    #     '9mos': ['M4', 'M5', 'M6', 'M7', 'M8', 'M9', 'M10', 'M11', 'M12'],
    #     '12mos': ['M1', 'M2', 'M3', 'M4', 'M5', 'M6', 'M7', 'M8', 'M9', 'M10', 'M11', 'M12']
    # }
    #  'M13_M24': {
    #      '3mos': ['M22', 'M23', 'M24'],
    #      '6mos': ['M19', 'M20', 'M21', 'M22', 'M23', 'M24'],
    #      '9mos': ['M16', 'M17', 'M18', 'M19', 'M20', 'M21', 'M22', 'M23', 'M24'],
    #      '12mos': ['M13', 'M14', 'M15', 'M16', 'M17', 'M18', 'M19', 'M20', 'M21', 'M22', 'M23', 'M24']
    # }
    #  'M25_M36': {
    #      '3mos': ['M34', 'M35', 'M36'],
    #      '6mos': ['M31', 'M32', 'M33', 'M34', 'M35', 'M36'],
    #      '9mos': ['M28', 'M29', 'M30', 'M31', 'M32', 'M33', 'M34', 'M35', 'M36'],
    #      '12mos': ['M25', 'M26', 'M27', 'M28', 'M29', 'M30', 'M31', 'M32', 'M33', 'M34', 'M35', 'M36']
    # }
    #  'M37_M48': {
    #      '3mos': ['M46', 'M47', 'M48'],
    #      '6mos': ['M43', 'M44', 'M45', 'M46', 'M47', 'M48'],
    #      '9mos': ['M40', 'M41', 'M42', 'M43', 'M44', 'M45', 'M46', 'M47', 'M48'],
    #      '12mos': ['M37', 'M38', 'M39', 'M40', 'M41', 'M42', 'M43', 'M44', 'M45', 'M46', 'M47', 'M48']
    # }
    #  'M49_M60': {
    #      '3mos': ['M58', 'M59', 'M60'],
    #      '6mos': ['M55', 'M56', 'M57', 'M58', 'M59', 'M60'],
    #      '9mos': ['M52', 'M53', 'M54', 'M55', 'M56', 'M57', 'M58', 'M59', 'M60'],
    #      '12mos': ['M49', 'M50', 'M51', 'M52', 'M53', 'M54', 'M55', 'M56', 'M57', 'M58', 'M59', 'M60']
    # }
     'M61_M72': {
         '3mos': ['M70', 'M71', 'M72'],
         '6mos': ['M67', 'M68', 'M69', 'M70', 'M71', 'M72'],
         '9mos': ['M64', 'M65', 'M66', 'M67', 'M68', 'M69', 'M70', 'M71', 'M72'],
         '12mos': ['M61', 'M62', 'M63', 'M64', 'M65', 'M66', 'M67', 'M68', 'M69', 'M70', 'M71', 'M72']
    }
    #  'M73_M84': {
    #      '3mos': ['M82', 'M83', 'M84'],
    #      '6mos': ['M79', 'M80', 'M81', 'M82', 'M83', 'M84'],
    #      '9mos': ['M76', 'M77', 'M78', 'M79', 'M80', 'M81', 'M82', 'M83', 'M84'],
    #      '12mos': ['M73', 'M74', 'M75', 'M76', 'M77', 'M78', 'M79', 'M80', 'M81', 'M82', 'M83', 'M84']
    # }
    #  'M85_M96': {
    #      '3mos': ['M94', 'M95', 'M96'],
    #      '6mos': ['M91', 'M92', 'M93', 'M94', 'M95', 'M96'],
    #      '9mos': ['M88', 'M89', 'M90', 'M91', 'M92', 'M93', 'M94', 'M95', 'M96'],
    #      '12mos': ['M85', 'M86', 'M87', 'M88', 'M89', 'M90', 'M91', 'M92', 'M93', 'M94', 'M95', 'M96']
    # }
    # 'M97_M108': {
    #     '3mos': ['M106', 'M107', 'M108'],
    #     '6mos': ['M103', 'M104', 'M105', 'M106', 'M107', 'M108'],
    #     '9mos': ['M100', 'M101', 'M102', 'M103', 'M104', 'M105', 'M106', 'M107', 'M108'],
    #     '12mos': ['M97', 'M98', 'M99', 'M100', 'M101', 'M102', 'M103', 'M104', 'M105', 'M106', 'M107', 'M108']
    # }
    # 'M109_M120': {
    #     '3mos': ['M118', 'M119', 'M120'],
    #     '6mos': ['M115', 'M116', 'M117', 'M118', 'M119', 'M120'],
    #     '9mos': ['M112', 'M113', 'M114', 'M115', 'M116', 'M117', 'M118', 'M119', 'M120'],
    #     '12mos': ['M109', 'M110', 'M111', 'M112', 'M113', 'M114', 'M115', 'M116', 'M117', 'M118', 'M119', 'M120']
    # }
}

# Define DPD thresholds
dpd_thresholds = [1, 15, 31, 61]

# Store results for merging
df_results = []

# Process only M1-M12
for period_group, month_groups in month_groups_dict.items():
    df_temp = df_unique_base.copy()
    
    for period, months in month_groups.items():
        df_filtered = df[df['MONTH'].isin(months)]
        
        # Compute delinquency flags for each threshold
        for dpd in dpd_thresholds:
            col_name = f'EverAtLeast{dpd}DPD_{period}'
            delq_flag = df_filtered.groupby('ILS_ACCT_NUM')['DELQ_AGE_ACTUAL'].apply(lambda x: int((x >= dpd).any()))
            df_temp[col_name] = df_temp['ILS_ACCT_NUM'].map(lambda x: delq_flag.get(x, 0))

        # Compute max WORST_DELINQUENCY per account
        worst_delq_max = df_filtered.groupby('ILS_ACCT_NUM')['WORST_DELINQUENCY'].max()
        df_temp[f'WorstDel_{period}'] = df_temp['ILS_ACCT_NUM'].map(lambda x: worst_delq_max.get(x, 0))

        # Count occurrences of Full, Partial, and Missed payments
        full_count = df_filtered[df_filtered['ACTUAL_AMORT_RATIO'] > 0.95].groupby('ILS_ACCT_NUM').size()
        partial_count = df_filtered[(df_filtered['ACTUAL_AMORT_RATIO'] > 0.10) & (df_filtered['ACTUAL_AMORT_RATIO'] <= 0.95)].groupby('ILS_ACCT_NUM').size()
        missed_count = df_filtered[df_filtered['ACTUAL_AMORT_RATIO'] <= 0.10].groupby('ILS_ACCT_NUM').size()

        df_temp[f'Full_past{period}'] = df_temp['ILS_ACCT_NUM'].map(lambda x: full_count.get(x, 0))
        df_temp[f'Partial_past{period}'] = df_temp['ILS_ACCT_NUM'].map(lambda x: partial_count.get(x, 0))
        df_temp[f'Missed_past{period}'] = df_temp['ILS_ACCT_NUM'].map(lambda x: missed_count.get(x, 0))

    # Assign Remaining Loan Ratio for M12 only
    df_m12 = df[df['MONTH'] == 'M72'][['ILS_ACCT_NUM', 'REMAINING_LOAN_RATIO']].set_index('ILS_ACCT_NUM')
    df_temp['RemainingLoanRatio_12mos'] = df_temp['ILS_ACCT_NUM'].map(df_m12['REMAINING_LOAN_RATIO'])

    df_results.append(df_temp)

# Combine results into a single DataFrame
df_unique = pd.concat(df_results, ignore_index=True)

# Display updated df_unique
df_unique.head(20)
```

## CREATE DEFAULT_TAGGING COL
