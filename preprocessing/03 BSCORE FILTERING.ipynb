```python
import pandas as pd
import calendar
```

```python
# Open Bscore

# Load the CSV file into a DataFrame
df_bscore = pd.read_csv('/Users/queencyangela/Desktop/M AMF/Second Sem/MATH 298.1/bscore_final.csv')  # Replace with your actual file name

# Display the first few rows
print(df_bscore)
```

```python
# Load the CSV file into a DataFrame
df_offset = pd.read_csv('/Users/queencyangela/Desktop/M AMF/Second Sem/MATH 298.1/Metrohome ils offset M97-120.csv')  # Replace with your actual file name

# Display the first few rows
print(df_offset)
```

```python
# Identify columns in df_bscore that contain "ACTUAL_AMORT_RATIO"
amort_ratio_cols = [col for col in df_bscore.columns if "ACTUAL_AMORT_RATIO" in col]

# Create df_amort with ILS_ACCT_NUM and ACTUAL_AMORT_RATIO columns
df_amort = df_bscore[['ILS_ACCT_NUM'] + amort_ratio_cols]

# Print the new DataFrame
print(df_amort)
```

```python
df_amort_cols = df_amort.columns

# Function to convert "ACTUAL_AMORT_RATIO MONTH YEAR" to "MM/DD/YYYY"
def rename_actual_amort_ratio(col_name):
    if "ACTUAL_AMORT_RATIO" in col_name:
        parts = col_name.split()  # Split by space
        if len(parts) == 3:  # Ensure it's in the expected format
            month, year = parts[-2], parts[-1]  # Extract month and year
            try:
                month = int(month)  # Convert month to integer
                year = int(year)    # Convert year to integer
                last_day = calendar.monthrange(year, month)[1]  # Get last day of month
                return f"{month:02d}/{last_day}/{year}"  # Format as MM/DD/YYYY
            except ValueError:
                return col_name  # Keep unchanged if conversion fails
    return col_name  # Keep other columns unchanged

# Apply renaming function to all column names
df_amort.columns = [rename_actual_amort_ratio(col) for col in df_amort.columns]

df_amort = df_amort[df_amort['ILS_ACCT_NUM'].isin(df_offset['ILS_ACCT_NUM'])]

# Merge df_amort with df_offset to include ORIG_BOOK_DATE
df_amort = df_offset.merge(df_amort, how='left', on='ILS_ACCT_NUM')

# Print the updated DataFrame
print(df_amort)
```

```python
# Convert DATE column to datetime (optional, only needed if date format varies)
df_amort['DATE_COLUMN'] = pd.to_datetime(df_amort['DATE_COLUMN'])

# Apply function to find the value in the matching date column
df_amort['ACTUAL_AMORT_RATIO'] = df_amort.apply(lambda row: row[row['DATE_COLUMN'].strftime('%m/%d/%Y')], axis=1)

# Drop all date columns dynamically (columns that match date format)
date_cols = df_amort.columns[9:-1]  # All columns after 'DATE' and before 'ACTUAL_AMORT_RATIO'
df_amort.drop(columns=date_cols, inplace=True)

print(df_amort)
```

```python
loan_ratio_cols = [col for col in df_bscore.columns if "REMAINING_LOAN_RATIO" in col]

df_loan = df_bscore[['ILS_ACCT_NUM'] + loan_ratio_cols]

# Print the new DataFrame
print(df_loan)
```

```python
# Convert DATE column to datetime (optional, only needed if date format varies)
df_loan['DATE_COLUMN'] = pd.to_datetime(df_loan['DATE_COLUMN'])

# Apply function to find the value in the matching date column
df_loan['REMAINING_LOAN_RATIO'] = df_loan.apply(lambda row: row[row['DATE_COLUMN'].strftime('%m/%d/%Y')], axis=1)

# Drop all date columns dynamically (columns that match date format)
date_cols = df_loan.columns[10:-1]  # All columns after 'DATE' and before 'ACTUAL_AMORT_RATIO'
df_loan.drop(columns=date_cols, inplace=True)

print(df_loan)
```
